```python
class Person():
    def __init__(self,n:str):
        self.name = n
    
    def __repr__(self):
        return f"這是Person的實體,我的名字是{self.name}" #字串插補
    
class Student(Person):
    def __init__(self,name:str,scores:int):
        super().__init__(name)
        self.scores = scores

s1 = Student("徐國堂",96)
print(type(s1))
print(s1.name)
print(s1.scores)
```
